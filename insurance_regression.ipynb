Simple Linear Regression Analysis

This script demonstrates the application of simple linear regression, focusing on the relationship
between age and BMI in the 'insurance.csv' dataset. The primary objective is to showcase the usage
of machine learning tools from the scikit-learn library.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
#load the dataset from local
local_path = "/Users/paul/Desktop/python school/T18 - Supervised learning - Simple Linear Regression/insurance.csv"
insurance_data = pd.read_csv(local_path)
insurance_data.head()
```

```python
# Check how data set in structured
insurance_data.info()

# Summary statistics
insurance_data.describe()
```

```python
# Plot a scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(insurance_data['age'], insurance_data['bmi'], alpha=0.8, color='green', label = 'Data Points')
# Adding the title and label
plt.title('Scatter Plot of Age vs BMI')
plt.xlabel('Age')
plt.ylabel('BMI')
# Adding grid lines
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
# Show the plot
plt.show()
```
